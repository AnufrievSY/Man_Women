```python
import dataframe
import pandas as pd
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
import numpy as np
df = pd.read_csv('МЖ.csv')
import time
```

```python
def find_index(version, value):
    start = time.time()
    r_v = []
    r_c = []
    
    if version == 1:
        for i in df.index:
            result = df.loc[i].values

            result = process.extractOne(v, result)
            r_v.append(i)
            r_c.append(result[1])
            
    elif version ==2:
        for i in df.index:
            result = df.loc[i].values
            for j in result:
                r = fuzz.WRatio(v, j)
                r_v.append(i)
                r_c.append(r)
                
    values = np.array(r_c)
    searchval = max(r_c)
    list(np.where(values == searchval)[0])
    print(time.time() - start)
    result = df.loc[[r_v[i] for i in np.where(values == searchval)[0]]]
    return result
```
