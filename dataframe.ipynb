```python
clean_df = df.loc[df['Суммарное кол-во просмотров'] > 1000000].sort_values('Суммарное кол-во просмотров')
clean_df['Название'].count()
text = ' '.join(clean_df['Описание'].to_list())
for t in [' ', '?', '!', '.', ',', ':']:
    text = text.replace(t, ' ')
pop = pd.DataFrame(text.split(' ')).value_counts()[1:]
pop = pop.loc[pop > 3]

for i in pop.index:
    i = i[0]
    if len(i)<4:
        pop = pop.drop(axis=0, index=i, level=None)
pop
```
